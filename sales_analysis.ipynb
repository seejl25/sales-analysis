```python
## Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Changing type of timestamp to timestamp data type
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Changing category to be of type string
df['category'] = df['category'].astype(str)

# Changing brand to be of type string
df['brand'] = df['brand'].astype(str)

# Changing user_id to be of type string
df['user_id'] = df['user_id'].astype(str)

# changing item_id to be of type string
df['item_id'] = df['item_id'].astype(str)
```

```python
# Checking for duplicates
df.duplicated().sum()
```

```python
# checking for null values
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# overall rating of all products
df['rating'].value_counts()
```

```python
# Visualising overall rating of all products
import seaborn as sns
sns.countplot(x='rating', data=df, palette="Set2")
plt.show()
```

From the above, we can see that most of the products sold have 5.0 ratings

```python
# Finding the year with the best sales
df['year'].value_counts()
```

```python
# Visualizing the year with the best sales
plt.figure(figsize=(15,7))
sns.countplot(x='year', data=df, palette="husl")
plt.show()
```

From the above graph, we can see that:
1. Sales in 2015 is the highest
2. There is an increasing trend of sales from 1999 to 2015 but drastically declined from 2015 to 2018

Seems weird that there was such as drastic drop from 2015 to 2016. Going back to the csv file, I realised that the year column and the year in the timestamp column do not match. For e.g in timestamp, a product was registered with the timestamp '1999-10-17', but the year column showed '2000'

```python
# Modifying the year column
df['year'] = df['timestamp'].dt.year
df.head(20)
```

```python
df.tail(10)
```

```python
# Since we have verified that the year column are now corresponding to the timestamp column, let's try plotting the graph again

plt.figure(figsize=(15,7))
sns.countplot(x='year', data=df, palette="husl")
plt.show()
```

After modifying the year column, we can see that:
1. 2015 still has the highest sales
2. There is still a drop from 2015 to 2018 but the drop is less drastic as compared to before

```python
# Generating the records for year 2015
df_2015 = df[df['year']==2015] 
df_2015
```

```python
# Determining which brand sold the most in 2015
df_2015['brand'].value_counts().iloc[1:10]
```

```python
# visualising the brands that sold the most in 2015
plt.figure(figsize=(10,5))
sns.countplot(data=df_2015, x='brand', order=df_2015['brand'].value_counts().iloc[1:10].index, palette='pastel')
plt.show()
```

From the graph, we can see the Bose and Mpow sold the most in 2015, with logitech close behind in third

```python
# Creating a new column to determine which month in 2015 had the most sales
df_2015['month'] = df_2015['timestamp'].dt.month_name()
df_2015.head(10)
```

```python
df_2015.tail(10)
```

```python
df_2015['month'].value_counts()
```

```python
# Visualising which month had the most sales in 2015
plt.figure(figsize=(12,5))
sns.countplot(data=df_2015, x='month',palette="pastel")
plt.show()
```

From the graph, we can see that August and December has the most sales in year 2015

```python
# Determining which category sold the most in 2015
df_2015['category'].value_counts()
```

```python
# Visualising which categories sold the most in 2015
plt.figure(figsize=(15,8))
plt.pie(data=df_2015, x=df_2015['category'].value_counts(), labels=df_2015['category'].unique())
plt.legend(title="Category", bbox_to_anchor=(1,0.7))
plt.show()
```

We can see that headphones are the most popular products sold in 2015
